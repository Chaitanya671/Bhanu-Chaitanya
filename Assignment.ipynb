```python
#From above columns we select the below features are the selected based on null values and business scenarios
```

```python
columns = ['fraudulent','orders/0/orderAmount','orders/0/orderState',
                 'paymentMethods/0/paymentMethodRegistrationFailure','paymentMethods/0/paymentMethodType',
            'transactions/0/transactionAmount',
                 'transactions/0/transactionFailed']
```

```python
df = df[columns]
```

```python
df.shape
```

```python
df.info()
```

```python
# Handling null values for continuous data
df['orders/0/orderAmount'].fillna(df['orders/0/orderAmount'].mean(),inplace=True)
df['transactions/0/transactionAmount'].fillna(df['transactions/0/transactionAmount'].mean(),inplace=True)
```

```python
# Handling null values for categorical data
for column in df.columns:
    for dtype in df.dtypes:
        if dtype=='object':
            df[column].fillna(df[column].mode()[0],inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
# Doing one hot encoding for categorical variables
df  = pd.get_dummies(df,columns = ['orders/0/orderState','paymentMethods/0/paymentMethodType'])
```

```python
# Label Encoding for bolean features 
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
```

```python
df['fraudulent'] = label_encoder.fit_transform(df['fraudulent'])
df['paymentMethods/0/paymentMethodRegistrationFailure'] = label_encoder.fit_transform(df['paymentMethods/0/paymentMethodRegistrationFailure'])
df['transactions/0/transactionFailed'] = label_encoder.fit_transform(df['transactions/0/transactionFailed'])
```

```python
df.head()
```

```python
X = df.loc[:, df.columns != 'fraudulent']
y = df['fraudulent']
```

# Model Building

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=1)
```

```python
# This is a Classification problem so we can test with Logistic Regression, Random Forest and Gradient Boosting
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr  = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
lr_predictions = lr.predict(X_test)
```

```python
print(confusion_matrix(y_test,lr_predictions))
print(accuracy_score(y_test,lr_predictions))
print(classification_report(y_test,lr_predictions))
```

# Random Forest Using for Feature Importence

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train,y_train)
```

```python
rf_predictions = rf.predict(X_test)
```

```python
print(confusion_matrix(y_test,rf_predictions))
print(accuracy_score(y_test,rf_predictions))
print(classification_report(y_test,rf_predictions))
```
